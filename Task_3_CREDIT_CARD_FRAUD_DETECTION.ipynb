#Split The Data Into Inputs Data And Target Data

```python
inputs_data = creditcard[creditcard.columns[:-1]]
target = creditcard['Class']
```

#Scaling / Normalize:


*   Normalize it means scaling the numerical columns it means values of occure in range dont Effect of outliers



```python
scaler = StandardScaler()
normalize_data = scaler.fit_transform(inputs_data)
```

#Split The Dataset Into Train And Test datasets

```python
x_train, x_test, y_train, y_test = train_test_split(normalize_data, target, test_size=0.2, random_state=5)
```

##Handle the Imbalance Class by using SMOTE Oversampling Technique
**SMOTE(Synthetic Minority Over-sampling Technique)**:

*   This technique generates synthetic samples for the minority class by interpolating between existing samples. It helps in mitigating the class imbalance issue.



```python
smote = SMOTE()
x_train, y_train = smote.fit_resample(x_train, y_train)
```

#Building LogisticRegression Model
  Trainig and evaluating of Logistic Regression

```python
reg_model = LogisticRegression()
reg_model.fit(x_train, y_train)
y_pred = reg_model.predict(x_test)
```

#Evalute the Accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print("accuracy_score{}".format(accuracy))
print("*******************************")

cr = classification_report(y_test, y_pred)
print("Classification Report:")
print(cr)
```

#Confusion Metrix

```python
print("Confusion Metrix:")
print(confusion_matrix(y_test, y_pred))
```

#Build RandomForest Classifier

```python
# Train RandomForest Classifier
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
y_pred1 = clf.predict(x_test)
```

```python
# Check Accuracy
accuracy1 = accuracy_score(y_test, y_pred1)
print("accuracy_score{}".format(accuracy1))
print("*******************************")

cr_clf = classification_report(y_test, y_pred1)
print("Classification Report:")
print(cr_clf)
```
